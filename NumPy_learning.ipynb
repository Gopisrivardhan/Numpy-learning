**Reorganizing arrays**

```python
bef = np.array([[1,2,3,4],[5,6,7,8]])
print(bef)
af = bef.reshape((2,2,2))
print(af)
```

```python
#vertical staacking
c1 = np.array([1,2,3,4])
c2 = np.array([5,6,7,8])
np.vstack([c1,c2,c2,c1])
```
